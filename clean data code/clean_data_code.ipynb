```python
import pandas as pd

df = pd.read_csv('/data/train.csv')

#xử lí dòng bị trùng
df.drop_duplicates(inplace=True)

#chuẩn hoá định dạng
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')

#ép kiểu Stt, sửa các giá trị bắt đầu bằng T là 1
df['Stt'] = df['Stt'].astype(str).str.replace(r'^T', '1', regex=True)
df['Stt'] = df['Stt'].astype(int)
df.drop_duplicates(inplace=True)

#làm tròn số thập phân rồi chuyển thành int
# thay , thành . ở những giá trị có chứa ,
df['Cost Amount'] = df['Cost Amount'].astype(str).apply(
    lambda x: x.replace(',', '.') if ',' in x else x
)
df['Cost Amount'] = df['Cost Amount'].astype(float)
df['Cost Amount'] = df['Cost Amount'].round()
df['Cost Amount'] = df['Cost Amount'].astype(int)
```
